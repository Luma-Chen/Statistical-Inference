# **1- Média amostral e variância amostral**

```python
%%R
mu<-2
sigma<-2
n<-10
asim<-10
xbar<-rep(NA, asim)
xvar<-rep(NA, asim)
for (i in 1:asim)
{
  print(i)
  set.seed(i)
  x<-rnorm(n, mu, sigma)
  xbar[i]<-mean(x)
  print(xbar[i])
  xvar[i]<-var(x)
  print(xvar[i])
}
```

(i) A média amostral é um estimador não viesado da média populacional, pois o valor esperado dela é sempre igual ao da média populacional. Além disso, as médias calculadas no código são sempre próximas da média populacional, mu= 2, variando por conta do desvio padrão.

(ii) Sim, a variância amostral é um estimador
não enviesado para a variância populacional. Porque o resultado esperado da variância amostral é igual ao da variância populacional. No caso, o resultado do código acima apresentou valores próximos da variância populacional, que é 4, ou seja, sendo 4 o resultado esperado de cada variância amostral.
